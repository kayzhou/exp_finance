```python
import numpy as np
import json

data = json.load(open('data/20180201-count-time-result.txt'))
# data

ts_clocks = np.zeros((16, 10))
amt_clocks = np.zeros((16, 10))

for d in data:
    dt = d[0]
#     print(dt.split('-'))
    if dt.split('-')[1] == 'None':
        continue
        
    clo = int(dt.split('-')[1]) - 1
    ts = d[1]['times']
    amt = d[1]['amt']
    
    ts_clocks[clo] += ts
    amt_clocks[clo] += amt

print(ts_clocks)
print(amt_clocks)
```

```python
import matplotlib.pyplot as plt

prop_0_ts = [ts_clocks[i][0] / sum(ts_clocks[i]) for i in range(ts_clocks.shape[0])]
X = range(len(prop_0_ts))
plt.bar(X, prop_0_ts)
plt.show()
```

```python
prop_0_amt = [amt_clocks[i][0] / sum(amt_clocks[i]) for i in range(amt_clocks.shape[0])]
X = range(len(prop_0_amt))
plt.bar(X, prop_0_amt)
plt.show()
```
